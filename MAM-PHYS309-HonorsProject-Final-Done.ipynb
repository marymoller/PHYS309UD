When first starting to learn relativity, you run into a snag when there is more than one observer.  It starts to look like you can actually bypass the so-called 'speed limit of the universe' and observe speeds **faster** than the speed of light.  In this tutorial, we're going to look at velocity addition in relativity, how we can adjust the classical Galilean transformations into something better suited for relativistic calculations, and how these adjustments affect the rest of the system.

### Why Not Galilean?

To understand why the Galilean transformations does not fully work in relativistic scenarios, let's set the scene.

We have two observers.  Frank is fixed, not moving, in frame K, and shoots a super powerful potato gun in the -x direction.  This is in the complete opposite direction of Mary, who is moving with frame K' in the +x direction, and passing Frank on her way.  If calculating the velocity of the potato using the Galilean transformation from her position moving away from it in frame K', Mary might find that that potato is going faster than the speed of light.

But mathematically, how does this relationship between $\vec{u}$ (velocity measured in frame K) and $\vec{u'}$ (velocity measured in frame K') present itself mathematically?

#### Time Transformation:

Galilean expression:
\begin{equation}
t=t'
\end{equation}
Applying transformation to differential form:
\begin{equation}
\frac{dt'}{dt}=1
\end{equation}

#### x Velocity Transformation:

Galilean expression:
\begin{equation}
x'=x-vt
\end{equation}
Applying transformation to differential form:
\begin{equation}
u'_x=\frac{dx'}{dt'}=\frac{dx'}{dt}\frac{dt}{dt'}=(\frac{dx}{dt}-v)(1)
\end{equation}
\begin{equation}
u'_x=u_x-v
\end{equation}

#### y Velocity Transformation:

Galilean expression:
\begin{equation}
y'=y
\end{equation}
Applying transformation to differential form:
\begin{equation}
u'_y=\frac{dy'}{dt'}=\frac{dy'}{dt}\frac{dt}{dt'}=(\frac{dy}{dt})(1)
\end{equation}
\begin{equation}
u'_y=u_y
\end{equation}

#### z Velocity Transformation:

Galilean expression:
\begin{equation}
z'=z
\end{equation}
Applying transformation to differential form:
\begin{equation}
u'_z=\frac{dz'}{dt'}=\frac{dz'}{dt}\frac{dt}{dt'}=(\frac{dz}{dt})(1)
\end{equation}
\begin{equation}
u'_z=u_z
\end{equation}

This gives us the existing, non-relativistic result we would expect.  However, now, we have to worry about the relativistic aspect of motion between the two frames, leading us to look into a set of adjusted transformations called the Lorentz transformations! (These can be solved for individually from the Galilean transformations, where you deduce the formula for $\gamma$, which I highly recommend!)

### Why Yes to Lorentz?

$\gamma$ moving forward is related to the velocity between frames, not the velocity of the potato being thrown.  This means that the v in the denominator is the velocity between the frames, not a transformed velocity in any direction.
\begin{equation}
\gamma = \frac{1}{\sqrt{1-(\frac{v^2}{c^2})}}
\end{equation}
This $\gamma$ can be treated as a constant throughout these transformation!

The Lorentz transformations relate the position and time in two separate inertial frames, the motion between which could possibly break the laws of physics if not treated correctly, as we have seen thus far.

#### Lorentz Time Transformation
\begin{equation}
ct'=\gamma[ct-(\frac{v}{c})x]
\end{equation}
\begin{equation}
t'=\gamma[t-(\frac{v}{c^2})x]
\end{equation}

#### x Velocity Transformation
\begin{equation}
x'=\gamma(x-vt)
\end{equation}

#### y Velocity Transformation
\begin{equation}
y'=y
\end{equation}

#### z Velocity Transformation
\begin{equation}
z'=z
\end{equation}


**Signs** are super important when dealing with Lorentz transformations!

In our example, when $u_x>0$, the x-component of the potato's velocity measured by Frank is in the **same direction** as the relative motion of Mary's frame.  This means that not only is the potato traveling towards Mary, but Mary is also traveling towards the potato.  This causes an addition of velocities of the potato and the velocity between the frames, and as a result, Mary will see the potato going slightly **faster** than predicted by the **Galilean** transformation.  

When $u_x<0$, the x-component of the potato's velocity measured by Frank is in the **opposite direction** as the relative motion of Mary's frame.  This means that the potato is flying away from Mary as Mary also moves away. The velocities will then add to **less** than predicted by the **Galilean** transformation.

It's also important to note that to reverse these transformations (to find $t$ or $x$ as opposed to $t'$ or $x'$), you simply switch whether the variables have a prime or not and reverse the signs, adding instead of subtracting.  For example:
\begin{equation}
x=\gamma(x'+vt')
\end{equation}
These reversed transformations are actually reversing the reference frame that you are observing from, or the one that you view to be fixed.

### Worked Example
Frank and Mary are working on a space station and need to clear away some nearby asteroids.  Frank stays on the space station and Mary gets in a spaceship and heads in the +x direction towards the asteroids at speed $v=0.6c$.  She shoots the asteroids with a proton gun, the protons moving with speed $u'=0.99c$ in the x-direction in her reference frame.  What speed does Frank observe from the space station?

Here, we can assign Frank to the fixed reference frame K, as Mary is moving in his frame of reference, and Mary to the moving reference frame K'.
The spaceship speed is represented by v, the speed of the proton in frame K' (Mary's perspective) is u', and the speed of the proton in frame K (Frank's perspective) is u, which is what we are aiming to find.
We can also assume that all the motion is happening in purely the x-direction.

First we have to directly relate $u_x$ to $x$ and $t$.
\begin{equation}
u_x=\frac{dx}{dt}
\end{equation}
This is also true for $u'_x$:
\begin{equation}
u'_x=\frac{dx'}{dt'}
\end{equation}
To get the expression for $u_x$, we're going to have to put our transformations in terms of the differentials $dx$ and $dt$, where $x=dx$, $x'=dx'$, $t=dt$, and $t'=dt'$.
\begin{equation}
dx=\gamma(dx'+v dt')
\end{equation}
\begin{equation}
dt=\gamma(dt'+\frac{v dx'}{c^2})
\end{equation}
Plug into equation for $u_x$ found above.
\begin{equation}
u_x=\frac{\gamma(dx'+v dt')}{\gamma(dt'+\frac{v dx'}{c^2})}
\end{equation}
Divide numerator and denominator by $dt'$
\begin{equation}
u_x=\frac{\frac{dx'}{dt'}+v}{1+\frac{v}{c^2}\frac{dx'}{dt'}}
\end{equation}
Notice how this canceled out the two $dt'$'s originally present in the equation.  It also allows us to simplify further, as $u'_x=\frac{dx'}{dt'}$.
\begin{equation}
u_x=\frac{u'_x+v}{1+\frac{v}{c^2}u'_x}
\end{equation}
This is our final expression for $u_x$, the speed of the proton from Frank's reference frame.
Plugging in the numbers, we find that
\begin{equation}
u_x=0.997c
\end{equation}

As mentioned above, to find $u'_x$ (the speed of the proton from Mary's reference frame), you simply have to reverse the signs and switch which $u_x$ and $u'_x$ (or which speed you are measuring!).  In this example, this gives us
\begin{equation}
u'_x=\frac{u_x-v}{1-\frac{v}{c^2}u_x}
\end{equation}

### Adding Velocity in Perpendicular Directions

If any angle is added to this problem, it will add movement and velocity in another perpendicular direction to the mix (either y or z).
For example, if the proton gun let out the protons in the shape of a cone, there would be elements of direction in both the y and z coordinates.

One thing that can be said about velocity in perpendicular directions is that it is still affected by relativity, even though it is in a direction perpendicular to the relative motion of the reference frames.  This includes **time dilation**.  This difference in time between the two reference frames shows itself in how long we perceive an object to be in motion for and, consequently, how fast it's going.  We normally see this effect in directions parallel to the relative motion, but as stated before, it still stands for perpendicular directions as well.

As a reminder, proper time is the shortest length of time observed, and should be linked to only one reference frame.  The other observed times should be longer.

Therefore, if Mary is observing the proper time, Frank will be observing time pass more **slowly** for Mary than she does herself.  Whatever action occurs, Frank will observe it taking **longer** to reach a certain point than Mary does, whether that action is the proton gun from our example or something like a pencil dropping.  And oppositely, Mary will observe time pass more **quickly** for Frank, so she will observe an action he took taking a **shorter** amount of time to reach the point in question.

Another thing affected by relativity is how we perceive space.  This includes angles and lengths and so on, such as the observed angles of the trajectory of the protons from the proton gun in our example.

In the original frame of reference (Mary, or the moving frame), we will assume that the angle $\theta'$ is given.  With Frank, however, you will need to find the angle relative to his axes.  This can be done using the geometry of the system and the ratios of speeds with relation to his reference frame.

As we can see above, this can form two different triangles, which can be defined by the respective velocities they perceive the proton to have ($u_x$, $u_z$, $\vec{u}$ **or** $u'_x$, $u'_z$, $\vec{u'}$).  To find the angle in Frank's reference frame, we can set up both of these triangles and solve for $\theta$ using information gathered through Lorentz transformations and additional geometric calculations.

We know $u_x$, $u'_x$, and $\theta'$.  From there, we can determine $u'_z$ through a trig calculation, which will be needed to determine $u_z$ later.
\begin{equation}
tan(\theta')=\frac{|u'_z|}{|\vec{u'}|}
\end{equation}
\begin{equation}
|u'_z|=|\vec{u'}|tan(\theta')
\end{equation}
Next, we can determine $u_z$ through our Lorentz transformations.  Remember $dz=dz'$ and $dt=\gamma(dt'+\frac{vdx'}{c^2})$:
\begin{equation}
u_z=\frac{dz}{dt}
\end{equation}
Plugging in Lorentz transformations
\begin{equation}
u_z=\frac{dz'}{dt'\gamma(1+\frac{vdx'}{dt'c^2})}
\end{equation}
Simplifying with $u'_x=\frac{dx'}{dt'}$ and $u'_z=\frac{dz'}{dt'}$
\begin{equation}
u_z=\frac{u'_z}{\gamma(1+\frac{u_xv}{c^2})}
\end{equation}

From here, we can now determine $\theta$ using trigonometric relations.
\begin{equation}
cos(\theta)=\frac{|u_x|}{|\vec{u}|}
\end{equation}
\begin{equation}
\theta=arccos(\frac{|u_x|}{|\vec{u}|})
\end{equation}

This gives you the angle of the cone that Frank observes from his own fixed reference frame, impacted by relativity and time dilation in a direction perpendicular to the relative motion of the reference frames.
